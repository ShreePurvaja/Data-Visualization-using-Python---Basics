```python
# line plot using bokeh

from bokeh.plotting import figure, output_notebook, show
from bokeh.io import push_notebook

# Step 1: Display output in the notebook
output_notebook()

# Step 2: Define data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Step 3: Create figure and plot
fig = figure(title='Line plot example', x_axis_label='x', y_axis_label='y')
fig.line(x, y)

# Step 4: Show the figure in the notebook
show(fig)
```

```python
# horizontal bar using bokeh

from bokeh.plotting import figure, output_notebook, show

output_notebook()

fig = figure(title='Bar plot example', width=400, height=400)

# Horizontal bar plot
fig.hbar(y=[2,4,6], right=[1,2,3], height=1,left=0, color="pink", legend_label="Horizontal Bars")

# Show the plot
show(fig)
```

```python
# vertical bar using bokeh

from bokeh.plotting import figure, output_notebook, show

output_notebook()

fig = figure(title='Bar plot example', width=400, height=400)

# Horizontal bar plot
fig.vbar(x=[2,4,6], top=[1,2,3], width=0.5,bottom=0, color="pink", legend_label="Vertical Bars")

# Show the plot
show(fig)
```

```python
# patch plot

from bokeh.plotting import figure, output_notebook, show

output_notebook()

fig = figure(title='Patch plot example', width=400, height=400)

fig.patch(x=[1,2,3,4],y=[2,3.5,7],color="deeppink")

show(fig)
```

```python
# patches

from bokeh.plotting import figure, output_notebook, show

output_notebook()

xs=[[5,3,4],[2,4,3],[2,3,5,4]]
ys=[[6,4,2],[3,6,7],[2,4,7,8]]

fig = figure(title='Patch plot example', width=400, height=400)

fig.patches(xs,ys,color=['pink','lightblue','purple'])

show(fig)
```

```python
# scatter markers

from bokeh.plotting import figure, output_notebook, show

output_notebook()

fig = figure(title='Scatter plot example', width=400, height=400)

fig.scatter([1,2,3,4,5],[6,5,2,4,7],marker="diamond",size=20,color='purple')

show(fig)
```

```python
# vertical area plot

from bokeh.plotting import figure, output_notebook, show

output_notebook()

x=[1,2,3,4,5]
y1=[2,6,4,3,5]
y2=[1,4,2,2,3]

fig = figure(title='Area plot example', width=400, height=400)

fig.varea(x=x,y1=y1,y2=y2,color='pink')

show(fig)
```

```python
# horizontal area plot

from bokeh.plotting import figure, output_notebook, show

output_notebook()

y=[1,2,3,4,5]
x1=[2,6,4,3,5]
x2=[1,4,2,2,3]

fig = figure(title='Area plot example', width=400, height=400)

fig.harea(y=y,x1=x1,x2=x2,color='pink')

show(fig)
```

```python
# circle glyphs

from bokeh.plotting import figure ,output_notebook,show

plot=figure(width=300,height=300)

plot.circle(x=[1,2,3],y=[3,7,5],size=20,color='pink')
plot.circle_cross(x=[2,4,6],y=[5,8,9],size=20,color='purple',alpha=0.6,line_width=3,line_color='pink')
plot.circle_x(x=[5,7,2],y=[2,4,9],size=20,color='lightblue',alpha=0.5,line_width=2,line_color='purple')

show(plot)
```

```python
# rectangle plot

from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.quad(top=[2,3,4],bottom=[1,2,3],left=[1,2,3],right=[1.2,2.7,3.5],color='green')

show(fig)
```

```python
# arbitrary rectangle

from bokeh.plotting import figure,output_notebook,show
from math import pi

fig=figure(width=300,height=300)

fig.rect(x=[1,2,3],y=[1,2,3],width=0.2,height=40,color='pink',angle=pi/3,height_units='screen')

show(fig)
```

```python
# ellipse plotting
from bokeh.plotting import figure,output_notebook,show
from math import pi

fig=figure(width=300,height=300)

fig.ellipse(x=[1,2,3],y=[1,2,3],width=[0.2,0.3,0.1],height=40,color='green',angle=pi/3,height_units='screen')

show(fig)
```

```python
# rectangle oval and polygon

from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.rect(x=10,y=10,width=100,height=50,color='pink',width_units='screen',height_units='screen')

fig.square(x=2,y=3,size=80,color='purple')

fig.ellipse(x=7,y=6,width=20,height=10,fill_color=None,line_width=2)


show(fig)
```

```python
# arc plotting
from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.arc(x=[1,2,3],y=[1,2,3],radius=0.1,start_angle=0.4,end_angle=4.8,color="navy")

show(fig)
```

```python
# wedge plotting
from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.wedge(x=[1,2,3],y=[1,2,3],radius=0.2,start_angle=0.4,end_angle=4.8,color="navy",alpha=0.6,direction='clock')

show(fig)
```

```python
# annular wedge plotting - filled area
from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.annular_wedge(x=[1,2,3],y=[1,2,3],inner_radius=0.1,outer_radius=0.25,start_angle=0.4,end_angle=4.8,color="navy",alpha=0.6,direction='clock')

show(fig)
```

```python
# annulus plotting - filled rings
from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

fig.annulus(x=[1,2,3],y=[1,2,3],inner_radius=0.1,outer_radius=0.25,color="navy",alpha=0.6,)

show(fig)
```

```python
# bezier curve

from bokeh.plotting import figure,output_notebook,show

fig=figure(width=300,height=300)

x=2
y=2

xp02=x+0.4
xp01=x+0.1
xm01=x+0.1
yp01=y+0.2
ym01=y-0.2

fig.bezier(x0=x,y0=y,x1=xp02,y1=y,cx0=xp01,cy0=yp01,cx1=xm01,cy1=ym01,color='navy')

show(fig)
```
