```python
# scatter plot

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="ticks", color_codes=True)

height=[62,64,69,75,66,68,65,71,76,73]

weight=[120,136,148,175,137,165,154,172,200,187]

sns.scatterplot(x=height,y=weight)

plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

df.head()
```

```python
print(sns.get_dataset_names())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=df['total_bill'], y=df['tip'], data=df)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=df['total_bill'], y=df['tip'], data=df,hue=df['smoker'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

hue_colors={'Yes':'Red','No':'Green'}
sns.scatterplot(x=df['total_bill'], y=df['tip'], data=df,hue=df['smoker'],hue_order=['No','Yes'],palette=hue_colors)

plt.show()
```

```python
# countplot

import matplotlib.pyplot as plt
import seaborn as sns

hue_colors={'Male':'Purple','Female':'Pink'}
sns.countplot(x=df['smoker'],data=df,hue=df['sex'],palette=hue_colors)

plt.show()
```

```python
#relplot
#subgroup with col
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',col=df['smoker'])

plt.show()
```

```python
#subgroups with row
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',row=df['smoker'])

plt.show()
```

```python
#subgroups with row and col
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',col=df['smoker'],row=df['time'])

plt.show()
```

```python
#subgroups for the days of the week
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',col=df['smoker'],row=df['day'])

plt.show()
```

```python
#wrapping and ordering columns

import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',col=df['day'],col_wrap=2,col_order=['Thur','Fri','Sat','Sun'])

plt.show()
```

```python
#subgroups with hue and size

import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',hue=df['size'],size=df['size'])
```

```python
#subgroups with hue and size with alpha (transparency)

import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='scatter',hue=df['size'],size=df['size'],alpha=0.6)
```

```python
#Line plot

import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], data=df, kind='line')
```

```python
#Line plot with customization

import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x=df['total_bill'], y=df['tip'], col=df['day'],data=df,hue=df['day'],style=df['day'],dashes=True, kind='line')
```
