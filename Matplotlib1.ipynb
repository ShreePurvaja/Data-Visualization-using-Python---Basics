```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

```python
#Labelling plot

import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.xlabel('Prices')
plt.ylabel('Sales')
plt.title('Line Plot')
plt.show()
```

```python
#customizing Line plot

import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y,marker='o',color='purple',linestyle='--')
plt.xlabel('Prices')
plt.ylabel('Sales')
plt.title('Line Plot')
plt.show()
```

```python
#customizing  plot

import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]
z=[15,8,7,5,6]
plt.plot(x,y,marker='o',color='purple',linestyle='None')
plt.plot(y,z,marker='v',color='orange',linestyle='None')
plt.xlabel('Prices')
plt.ylabel('Sales')
plt.title('Line Plot')
plt.show()
```

```python
#customizing  plot

import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]
z=[15,8,7,5,6]
plt.plot(x,y,marker='o',color='purple',linestyle='None')
plt.plot(y,z,marker='v',color='orange',linestyle='--')
plt.xlabel('Prices')
plt.ylabel('Sales')
plt.title('Line Plot')
plt.show()
```

```python
#customizing  plot

import matplotlib.pyplot as plt

x=[5,2,9,4,7]
y=[10,5,8,4,2]
z=[15,8,7,5,6]
plt.plot(x,y,marker='o',color='purple',linestyle='None',label='XY')
plt.plot(y,z,marker='v',color='orange',linestyle='--',label="YZ")
plt.plot(x,z,color='g',label="XZ")
plt.xlabel('Prices')
plt.ylabel('Sales')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(2,1)

ax[0].plot(x,y,marker='o',color='purple',linestyle='--',label='XY')
ax[1].plot(y,z,marker='v',color='orange',linestyle='--',label="YZ")

ax[0].set_ylabel('Prices')
ax[1].set_ylabel('Sales')
ax[1].set_xlabel('Line Plot')

plt.show()
```

```python
import pandas as pd

df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

figsize=(20,10)
plt.figure(figsize=figsize)
df.plot(fontsize=10)
plt.legend(fontsize=10)
plt.show()
```

```python
df_subset = df.loc[20:49, ['Age', 'Outcome']]

ax=df_subset.plot(figsize=(12,5),fontsize=14,marker='o',color='purple',linestyle='--',label='Age')

ax.set_xlabel('Age')
ax.set_ylabel('Outcome')
ax.set_title('Age vs Outcome')
plt.show()
```

```python
df_subset = df.loc[20:49, ['Age', 'Outcome']]

ax=df_subset.plot(figsize=(12,5),fontsize=14,marker='o',color='purple',linestyle='--',label='Age')

ax.axvline(x=30,color='red')
ax.axvline(x=40,color='red')
ax.axvline(x=50,color='red')
ax.axvline(x=60,color='red')
ax.axhline(y=1,color='red')
plt.show()
```

```python
df_subset = df.loc[20:70, ['Age', 'Outcome']]

ax=df_subset.plot(figsize=(12,5),fontsize=14,marker='o',color='purple',linestyle='--',label='Age')

ax.axhspan(10,20,alpha=0.2,color='red')
ax.axvspan(30,40,alpha=0.2,color='red')

plt.show()
```
